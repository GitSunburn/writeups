```python
bits = 2048
RF = RealField(bits)
p = next_prime(round(2^(bits-1)*(1+RF.random_element(0,1))), proof=False)
q1 = next_prime(round(2^(bits-1)*(1+RF.random_element(0,1))), proof=False)
q2 = next_prime(round(2^(bits-1)*(1+RF.random_element(0,1))), proof=False)
e = 0x010001
```

```python
def generate_file(p, q, e):
    d = mod(1/e, (p-1)*(q-1))
    return """\
asn1=SEQUENCE:rsa_key

[rsa_key]
version=INTEGER:0
modulus=INTEGER:"""+str( p*q )+"""
pubExp=INTEGER:"""+str( e )+"""
privExp=INTEGER:"""+str( d )+"""
p=INTEGER:"""+str( p )+"""
q=INTEGER:"""+str( q )+"""
e1=INTEGER:"""+str( mod(d, p-1) )+"""
e2=INTEGER:"""+str( mod(d, q-1) )+"""
coeff=INTEGER:"""+str( mod(1/q, p) )+"""
"""
```

```python
open('genconf1', 'w').write(generate_file(p, q1, e))
```

```python
%%bash
rm key1.private.der key1.private.pem key1.cert.pem
openssl asn1parse -genconf genconf1 -out key1.private.der
openssl rsa -inform der -in key1.private.der -out key1.private.pem
openssl req -key key1.private.pem -x509 -days 365 -out key1.cert.pem -nodes
```
